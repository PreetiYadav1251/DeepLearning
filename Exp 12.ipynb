```python
# Import Library
from sklearn.feature_extraction.text import CountVectorizer

# Sample Text Data
documents = [
    "Natural language processing is interesting",
    "Machine learning is powerful",
    "Natural language processing and machine learning"
]

# Create CountVectorizer Object
vectorizer = CountVectorizer()

# Fit and Transform Data
X = vectorizer.fit_transform(documents)

# Print Vocabulary
print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nBag of Words Matrix:")
print(X.toarray())
```

```python
# Import Library
from sklearn.feature_extraction.text import CountVectorizer

# Sample Text Data
documents = [
    "Natural language processing is interesting",
    "Machine learning is powerful",
    "Natural language processing and machine learning"
]

# Create Bigram Model
vectorizer_ngram = CountVectorizer(ngram_range=(2,2))

# Fit and Transform
X_ngram = vectorizer_ngram.fit_transform(documents)

# Print Vocabulary
print("Bigram Vocabulary:")
print(vectorizer_ngram.get_feature_names_out())

print("\nBag of n-grams Matrix:")
print(X_ngram.toarray())
```

```python
# Import Library
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample Text Data
documents = [
    "Natural language processing is interesting",
    "Machine learning is powerful",
    "Natural language processing and machine learning"
]

# Create TF-IDF Vectorizer
tfidf = TfidfVectorizer()

# Fit and Transform
X_tfidf = tfidf.fit_transform(documents)

# Print Vocabulary
print("Vocabulary:")
print(tfidf.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(X_tfidf.toarray())
```
